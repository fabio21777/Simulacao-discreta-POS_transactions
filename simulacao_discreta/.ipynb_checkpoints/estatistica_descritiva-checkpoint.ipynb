```python
import numpy as np
import pandas as pd
import re
```

```python
trasacao = pd.read_csv("data/POS_transactions_20190328-20190410.csv")
```

```python
trasacao.head()
```

```python
trasacao.dtypes
```

```python
trasacao_workstation_service = trasacao.query('WorkstationGroupID == 1')
```

```python
trasacao_workstation_service_date = trasacao_workstation_service.loc[trasacao_workstation_service.BeginDateTime.str.contains('2019-04-03T10|2019-04-03T11|2019-04-03T12', flags=re.I, regex=True)]
```

```python
trasacao_workstation_service_date.describe()
```

```python
OperatorIDs=trasacao_workstation_service_date[['OperatorID']]
```

```python
operadores_unicos = OperatorIDs.drop_duplicates()
operadores_list = operadores_unicos.values
```

```python
operadores = {}
for operador in operadores_list:
    operadores[int(operador)] = trasacao_workstation_service_date.query(f"OperatorID=={operador}")
```

```python
trasacao_workstation_service_date_operador301 = trasacao_workstation_service_date.query("OperatorID == 301")
print("media",trasacao_workstation_service_date_operador301["TranTime"].mean())
print("mediana",trasacao_workstation_service_date_operador301["TranTime"].median())
print("Moda",list(trasacao_workstation_service_date_operador301["TranTime"].mode()))
print("minimo",trasacao_workstation_service_date_operador301["TranTime"].min())
print("maximo",trasacao_workstation_service_date_operador301["TranTime"].max())
print("amplitude",trasacao_workstation_service_date_operador301["TranTime"].max()-trasacao_workstation_service_date_operador301["TranTime"].min())
print("desvio padrão",trasacao_workstation_service_date_operador301["TranTime"].std())
print("variacia",trasacao_workstation_service_date_operador301["TranTime"].var())
print("coeficinte de variação",trasacao_workstation_service_date_operador301["TranTime"].std()/trasacao_workstation_service_date_operador301["TranTime"].mean())
print("Assimetria",trasacao_workstation_service_date_operador301["TranTime"].skew())
trasacao_workstation_service_date_operador301.describe()
```
